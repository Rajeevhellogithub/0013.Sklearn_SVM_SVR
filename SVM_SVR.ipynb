```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
dataset = pd.read_csv(r'E:\PYTHONCLASSJUPYTER\PrakashSenapati\2024_10_03_Svr\emp_sal.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
x = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
x[0:5]
```

```python
y[0:5]
```

```python
svr_regressor = SVR(kernel='poly', degree=5, gamma='scale')
svr_regressor.fit(x,y)
```

```python
knn_regressor = KNeighborsRegressor(n_neighbors=5, weights='distance', p=2)
knn_regressor.fit(x,y)
```

```python
dt_regressor = DecisionTreeRegressor(splitter='random', criterion='absolute_error')
dt_regressor.fit(x,y)
```

```python
rf_regressor = RandomForestRegressor(random_state=0, n_estimators=50)
rf_regressor.fit(x,y)
```

```python
y_pred_svr = svr_regressor.predict([[6.5]])
y_pred_svr
```

```python
y_pred_knn = knn_regressor.predict([[6.5]])
y_pred_knn
```

```python
y_pred_dt = dt_regressor.predict([[6.5]])
y_pred_dt
```

```python
y_pred_rf = rf_regressor.predict([[6.5]])
y_pred_rf
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, svr_regressor.predict(x), color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
